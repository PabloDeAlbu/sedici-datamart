### **Personal de ciencia y tecnología**

El personal de CyT en una unidad estadística incluye a todas las personas involucradas directamente en I+D así como a aquellas que brindan servicios directos para las actividades de I + D (como gerentes de I + D, administradores, técnicos y personal de oficina).

* **Página de referencia:** https://datasets.datos.mincyt.gob.ar/dataset/personal-de-ciencia-y-tecnologia



#### **personas**
Conjunto de datos que describe mediante atributos tales como nombre, apellido, sexo, edad, a las personas físicas.

### **Proyectos de ciencia, tecnología e innovación**

Son unidades estadísticas que incluyen los proyectos de I+D y/o de innovación (conjunto de actividades que se llevan a cabo para crear resultados CyT y/o innovativos en un tiempo determinado).

#### **Recurso 'personas_2011.csv'**

* **Frecuencia de actualización:** Dicen que es anualmente, pero han habido actualizaciones recientemente (hace 6 meses).
* **Fecha de publicación:** 8 de febrero de 2019
* **Página de referencia:** https://datasets.datos.mincyt.gob.ar/dataset/personal-de-ciencia-y-tecnologia

Persona se define como una entidad relacional descrita a través de atributos tales como: sexo, edad, nivel educativo y otros.

| Título de la columna | Tipo de dato | Descripción |
| ----------- | ----------- | ----------- |
| persona_id | integer | ID de la persona. Ver recurso Personas  |
| anio | integer | Año de referencia.  |
| sexo_id | integer | ID de sexo de la persona. Ver recurso ref_SEXO  |
| edad | integer | Edad de la persona en el año de referencia.  |
| maximo_grado_academico_id | integer | Id del máximo grado académico concluido por la persona hasta el año de referencia. Ver recurso ref_GRADO_ACADEMICO  |
| disciplina_maximo_grado_academico_id | integer | ID de la disciplina del máximo grado académico concluido alcanzado por la persona. Ver recurso ref_DISCIPLINA  |

| disciplina_experticia_id | integer | ID de la disciplina de experticia declarada por la persona. Ver recurso ref_DISCIPLINA  |
| tipo_personal_id | integer | Id del tipo de personal que corresponde a la persona en el año de referencia. Ver recurso ref_TIPO_PERSONAL  |
| producciones_ult_anio | integer | Total de producciones cyt de la persona en el año de referencia. Ver https://github.com/sicytar/ckan/wiki/producciones  |

| producciones_ult_4_anios | integer | Total de producciones cyt de la persona en el año de referencia y los tres anteriores. Ver https://github.com/sicytar/ckan/wiki/producciones  |
| institucion_trabajo_id | integer | ID de la institución de trabajo declarada por la persona. Ver recurso organizaciones  |
| seniority_level | string | Variable de “categoría de investigador”. Ver https://github.com/sicytar/ckan/wiki/seniority  |
